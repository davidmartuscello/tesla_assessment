```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest

import warnings
warnings.filterwarnings(action='ignore')
```

```python
import pickle
with open("persistent_dataframe",'rb') as f:
    site_dataframes = pickle.load(f)  
```

## Remove Sites With No Solar Power Data

```python
sites_with_no_solar_data = []
site_dataframes_with_solar = {}
for site_id, site_df in site_dataframes.items():
    if len(site_df['SolarPower'].values) == 0:
        sites_with_no_solar_data.append(site_id)
    else:
        site_dataframes_with_solar[site_id] = site_df

# These sites are outliers and it should be investigated why they are not providing data
sites_with_no_solar_data
```

## Outlier Detection Using Descriptive Statistics

The time-series nature of out data makes it harder to discern outliers. One way to get around this is to use descriptive statistics for each site as features to differentiate them. Each site gets one row in the 'solar_df' dataframe, with features like mean, std, min, etc.

```python
# Compile descriptive statistics into solar_df
solar_df = pd.DataFrame()
for site_id, site_df in site_dataframes_with_solar.items():
    stats_ser = pd.Series(site_df['SolarPower'].values).describe()
    stats_df = pd.DataFrame(stats_ser).transpose().drop(columns=['count', '25%', '50%', '75%'])
    stats_df['site'] = site_id
    solar_df = pd.concat([solar_df, stats_df])
    solar_df.reset_index(drop=True, inplace=True)
```

Once the data is loaded into dataframe, it can be analyzing with traditional anomaly detection models. In this case I used the IsolationForest model.

```python
outliers_fraction = 0.10
clf = IsolationForest(contamination=outliers_fraction, random_state=42)

X = solar_df.drop(columns=['site'])
y_if = clf.fit(X).predict(X)

colors = np.array(["#377eb8", "#ff7f00"])

fig, axs = plt.subplots(len(X.columns),figsize=(10,13))
fig.suptitle("SolarPower Across Sites Over Time")
for i, col_name in enumerate(X.columns):
    axs[i].scatter(solar_df['site'], X[col_name], s=25, color=colors[(y_if + 1) // 2])
    axs[i].set_ylabel(col_name)
    axs[i].set_xlabel('site')
    axs[i].set_xticklabels(X.index, rotation = 0)

fig.tight_layout()
```

```python
outliers_detected_from_descriptive_stats = [site_id for site_id, y in zip(solar_df['site'], y_if) if y == -1]
outliers_detected_from_descriptive_stats
```

## Principal Component Analysis

PCA is used here to compress the feature space so that we can clearly see which points are outliers to verify that our model is choosing well. 

```python
import numpy as np
from sklearn.decomposition import PCA
# X = next(iter(site_dataframes.values()))['SolarPower']
X = solar_df.drop(columns=['site'])
pca = PCA(n_components=2)
y_pca = pca.fit_transform(X)
y_pca = pd.DataFrame(y_pca, columns=['pc1', 'pc2'])
```

```python
outliers_fraction = 0.20
clf = IsolationForest(contamination=outliers_fraction, random_state=42)

X = y_pca
y_pca_if = clf.fit(X).predict(X)

colors = np.array(["#377eb8", "#ff7f00"])

fig, axs = plt.subplots(1,figsize=(5,5))
fig.suptitle("PCA of Descriptive Statistics")
axs.scatter(y_pca['pc1'], y_pca['pc2'], s=10, color=colors[(y_pca_if + 1) // 2])
axs.set_xlabel('pc1')
axs.set_ylabel('pc2')
axs.set_xticklabels(X.index, rotation = 0)
fig.tight_layout()
```
